Покажем, что вычисления на cpu могут быть быстрее, чем на gpu. Например, если тензоры для вычислений взять маленькими.

```python
import numpy as np
import torch
```

```python
%%time
a = torch.randint(0, 100, (1000,))
b = torch.randint(0, 100, (1000,))
print(a)
x = a * b
```

```python
%%time
a_t = torch.randint(0, 100, (1000,)).to('cuda:0')
b_t = torch.randint(0, 100, (1000,)).to('cuda:0')
print(a_t)
x_t =  (a_t * b_t).to('cpu')
```
